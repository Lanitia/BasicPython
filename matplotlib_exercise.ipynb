## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_hist = pd.read_csv("data/hist_data.csv")
plt.hist(df_hist["aデータ"], bins=30, alpha=0.5, color="y", label="aデータ")
plt.hist(df_hist["bデータ"], bins=30, alpha=0.5, color="c", label="bデータ")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.legend(loc="upper right")
plt.title("aデータとbデータの温度に関する度数分布")
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_scatter = pd.read_csv("data/scatter_data.csv")

class_0 = df_scatter[df_scatter["class"] == 0]
class_1 = df_scatter[df_scatter["class"] == 1]
class_2 = df_scatter[df_scatter["class"] == 2]
plt.scatter(class_0['length (cm)'], class_0['width (cm)'], color="r", label="class0")
plt.scatter(class_1['length (cm)'], class_1['width (cm)'], color="b",label="class1")
plt.scatter(class_2['length (cm)'], class_2['width (cm)'], color="g",label="class2")
plt.xlabel('length (cm)')
plt.grid(True)
plt.ylabel('width (cm)')
plt.legend()
plt.title('クラス０～２のlengthとwidthに関する散布図')
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df_line = pd.read_csv("data/line_data.csv")

#折れ線
plt.plot(df_line["期間"], df_line["件数"], marker="o")
plt.grid(True)
plt.xticks(rotation=30)
plt.xlabel("期間")
plt.ylabel("件数")
plt.title("期間ごとの件数の推移")
```

```python
#棒
plt.bar(df_line["期間"], df_line["費用"])
plt.grid(True)
plt.xticks(rotation=30)
plt.xlabel("期間")
plt.ylabel("費用（円）")
plt.title("期間ごとの費用の推移")
```
